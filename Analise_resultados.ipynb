# Graficos

Foi avaliado pelos modelos duas amostras de imagens com cada amostra possuindo 100 imagens, uma amostra com apenas folhas saudaveis e outra com apenas folhas doentes. \
Em cada amostra foi analizado quantos porcento dessas imagens o modelo classificou como sendo folhas saudaveis, folhas doentes e incerto. \
O modelo realizou 50 avaliações para a mesma amostra de imagens, sendo que em cada avaliação foi variado o nivel de precisão do modelo, variando de 50% a 100%. \
O que siginifica que com uma precisão de 95%, o modelo classificará uma imagem como saudavel ou doente se ele tiver 95% ou mais de confiança que se trata de uma folha saudavel ou doente, caso contrário ela sera classificada com incerto. \
Com isso foi gerado um conjunto de 6 gráficos, sendo 3 para a amostra de folhas saudaveis e 3 para a de folhas doentes. \
Para cada amostra foi ferado um gráfico para a taxa classificação das folhas em saudaveis, doentes e incertos. 



## Amostras Saudaveis

explicar a precisao de 100%

```python
# Filtra os DataSets com apenas os dados da Amostra saudavel 

# Xceptio 200
df_Xception_200_saudaveis = df_Xception_200[(df_Xception_200['Amostra'] == "saudavel")]
df_Xception_200_saudaveis = df_Xception_200_saudaveis.drop(50) # Remove a ultima linha, a que tem precisao de 100%


# Xceptio 450
df_Xception_450_saudaveis = df_Xception_450[(df_Xception_450['Amostra'] == "saudavel")]
df_Xception_450_saudaveis = df_Xception_450_saudaveis.drop(101) # Remove a ultima linha, a que tem precisao de 100%

# Xception pré treinada 40/20
df_Xception_pre_treined_saudaveis = df_Xception_pre_treined[(df_Xception_pre_treined['Amostra'] == "saudavel")]
df_Xception_pre_treined_saudaveis = df_Xception_pre_treined_saudaveis.drop(50) # Remove a ultima linha, a que tem precisao de 100%

# Xception pré treinada 80/40
df_Xception_pre_treined_saudaveis_80_40 = df_Xception_pre_treined_80_40[(df_Xception_pre_treined_80_40['Amostra'] == "saudavel")]
df_Xception_pre_treined_saudaveis_80_40 = df_Xception_pre_treined_saudaveis_80_40.drop(50) # Remove a ultima linha, a que tem precisao de 100%

# Xception pré treinada 100/60
df_Xception_pre_treined_saudaveis_100_60 = df_Xception_pre_treined_100_60[(df_Xception_pre_treined_100_60['Amostra'] == "saudavel")]
df_Xception_pre_treined_saudaveis_100_60 = df_Xception_pre_treined_saudaveis_100_60.drop(50) # Remove a ultima linha, a que tem precisao de 100%
```

### Visão Geral dos Resultados.

#### Xception 200 Epocs

```python
x = df_Xception_200_saudaveis['Precisao']
y = df_Xception_200_saudaveis['Saudaveis']
z = df_Xception_200_saudaveis['Doentes']
k = df_Xception_200_saudaveis['Incertos']

plt.title("Xception 200 Epocs (100 Amostras saudaveis)")
plt.xlabel("Precisão (%)")
plt.ylabel("Taxa de Classificação (%)")
plt.rcParams['figure.figsize'] = (14, 10)
plt.grid()
plt.yticks(np.arange(0, 100, 5))
plt.xticks(np.arange(50, 105, 5))
plt.plot(x,y, color= 'blue', label='Saudaveis')
plt.plot(x,z, color= 'red', label='Doentes')
plt.plot(x,k, color= 'green', label='Incerto')
plt.legend()
plt.show()
```
